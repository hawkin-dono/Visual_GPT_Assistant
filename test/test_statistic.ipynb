```python
df = pd.read_csv("../data/maintain.csv")
df.head()
```

```python

machine_name = "OP4"
machine_data = df[df['Tên thiết bị'] == machine_name]
```

```python
import matplotlib.pyplot as plt
```

```python

if machine_data.empty:
    raise ValueError(f"No data found for machine type {machine_name}")

cause_counts = machine_data['Mã Nguyên nhân'].value_counts()

cause_counts

```

```python
for i in cause_counts.index:
    tmp = machine_data.loc[machine_data['Mã Nguyên nhân'] == i]['Nguyên nhân 1']
    print(f"mã lỗi {i} tương đương với các nguyên nhân: {tmp.values} xuất hiện {cause_counts[i]} lần")
    
    
```

```python
plt.figure(figsize=(10, 6))
cause_counts.plot(kind='bar', color='skyblue', alpha=0.8)
plt.title(f"Common Causes for Machine Type: {machine_name}", fontsize=16)
plt.xlabel("Cause Code (Mã Nguyên nhân)", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
```
